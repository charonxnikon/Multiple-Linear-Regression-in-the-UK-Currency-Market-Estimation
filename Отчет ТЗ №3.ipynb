```r
nz <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/1.csv",sep = ';', header = TRUE, dec = ".")
pred1 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred1.csv", sep = ';', header = TRUE, dec = ".")
pred2 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred2.csv", sep = ';', header = TRUE, dec = ".")
pred3 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred3.csv", sep = ';', header = TRUE, dec = ".")
pred4 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred4.csv", sep = ';', header = TRUE, dec = ".")
pred5 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred5.csv", sep = ';', header = TRUE, dec = ".")
pred6 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred6.csv", sep = ';', header = TRUE, dec = ".")
pred7 <- read.csv2("/Users/Nikon/Desktop/CMC MSU/MC/6 sem/R/3/pred7.csv",sep = ';', header = TRUE, dec = ".")
```

Объединнеие разных баз данных в единную базу М и сохранизация их по слобцу "DATE"

```r
M <- merge(nz, pred1, by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred2,by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred3,by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred4,by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred5,by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred6,by = "X.DATE.", incomparables = NA, sort = FALSE) %>%
  merge(pred7,by = "X.DATE.", incomparables = NA, sort = FALSE)
M
```

Сохранение объединенной выборки средствами R через "save" в файл "Stocks.RData" и через "write.csv2" в файл "Stocks.csv"

```r
save(M, file = "Stocks.RData")
write.csv2(M, "Stocks.csv")
```

Переход к логарифмисеской доходности

```r
F_ <- diff(log(M$X.CLOSE.F.))
pred_1 <- diff(log(M$X.CLOSE.INDEX.))
pred_2 <- diff(log(M$X.CLOSE.GOLD.))
pred_3 <- diff(log(M$X.CLOSE.OIL.))
pred_4 <- diff(log(M$X.CLOSE.FRANK.))
pred_5 <- diff(log(M$X.CLOSE.EURO.))
pred_6 <- diff(log(M$X.CLOSE.PLATINUM.))
pred_7 <- diff(log(M$X.CLOSE.CROWN.))
```

Построение регрисионной модели в первом приближении 

```r
reg <- lm(F_ ~ pred_1 + pred_2 + pred_3 + pred_4 + pred_5 + pred_6 + pred_7)
summary(reg)
```

Определение значимых и незначимых предикторов

Гораздо удобнее определять через таблицу корреляции (corrplot) вместо F-test 

Такой подход был использован в прошлом семестре

```r
states <- as.data.frame(M[,c("X.CLOSE.F.", "X.CLOSE.INDEX.", "X.CLOSE.GOLD.","X.CLOSE.OIL.", "X.CLOSE.FRANK.", "X.CLOSE.EURO.", "X.CLOSE.PLATINUM.", "X.CLOSE.CROWN.")])
cor_table <- cor(states)
```

Задание градиента цветов

```r
col4 <- colorRampPalette(c("#7F0000","#FF7F00", "#7FFF7F", "#007FFF","#000000"))
corrplot(cor_table, method = "shade", col = col4(10), cl.length = 11, order = "AOE", addCoef.col = "red")
```

Незначимыми предикторами будем называть те, что близко к нулю(низкокоррелируемые), 
а значимыми те, что ближе к 1(коррелирумые) или -1 (обратно коррелируемые)

Отметим, что разницы строить таблицу корреляции по начальным данным или по логарифмической доходности нет

Построение регрисионной модели во втором приближении (для значимых предикторов)

```r
new_reg <- lm(F_ ~ pred_7 + pred_6 + pred_1 + pred_5)
```

Нахождение p-value для ответа на вопрос вырожденности модели

Ответ: обе модели невырожденные, поскольку p-value малое

```r
summary(new_reg)
summary(reg)
```

Если модель выбрана верно, то поведение процесса остатков п.н. адекватно предположениям относительно процесса "xt" перед оценкой модели
т.е если модель "xt" - Гауссовский шум -> модель "xˆt" тоже. 

Если это не так, модель тренда неверна и не учитывает всю специфику динамики 

Построение гистрограммы остатков

Очевидно видно, что близко к нормальному распределению

```r
hist(rstudent(new_reg), xlab = 'Standardized Residuals', col = "cadetblue1")
```

Тест Колмогорова-Смирнова для нормированных остатков

Ответ: Остатки не нормально распределенные

```r
ks.test(rstudent(new_reg), "pnorm", 0, 1)
```

Гипотеза о распределении Стьюдента -> Тест Пирсона для распределения Стьюдента на остатках

Ответ: остатки не имеют распределения Стьюдента

```r
pearson.test(rstudent(new_reg))
```

Построение нормального квантиль-квантиль графика (QQ графика)

Если распределение с.в. нормальное, то график - прямая 

На графике ниже концы не прямые -> есть тяжелые хвосты -> остатки п.н. не нормально 

```r
qqnorm(rstudent(new_reg),col = "cadetblue2")
```

Построение автокорреляционной полосы

Ответ: Почти все автокорреляции лежат внутри полосы -> остатки некоррелируют, не зависимы

```r
acf(rstudent(new_reg))
```
